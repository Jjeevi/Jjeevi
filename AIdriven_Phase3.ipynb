```python
import pandas as pd
import pandas as random
import pandas as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('C:\\Users\\win10\\Desktop\\Data_Gov_Tamil_Nadu.csv',encoding='latin-1')
pd.set_option("display.max.columns", None)
df.head()
```

```python
x=df['AUTHORIZED_CAP']
y=df['PAIDUP_CAPITAL']
plt.scatter(x,y)
plt.xlabel('AUTHORIZED_CAP')
plt.ylabel('PAIDUP_CAPITAL')
plt.title('Scatter plot')
plt.show()
```

```python
sentiment_counts = df['AUTHORIZED_CAP'].value_counts() 
labels = sentiment_counts.index 
sizes = sentiment_counts.values
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140) 
plt.axis('equal') 
plt.title("Sentiment Distribution",color='#a114de') 
plt.show()
```
